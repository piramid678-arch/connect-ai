```python
import yaml

# 8강 원리에 따른 합치기 설정 (SLERP 방식으로 50:50 섞기)
merge_config = """
merge_method: slerp
base_model: "HuggingFace_베이스_모델_주소" # 예: google/gemma-1.1-2b-it
models:
  - model: "HuggingFace_베이스_모델_주소"
  - model: "내가_만든_혹은_특화된_모델_주소"
parameters:
  t: 0.5  # 0=A쪽, 1=B쪽, 0.5=반반 섞기
dtype: float16
"""

# 설정 파일을 config.yaml로 저장
with open("config.yaml", "w") as f:
    f.write(merge_config)

print("✅ config.yaml 병합 지시서 생성 완료")
```

```python
config = '''
slices:
  - sources:
      - model: Qwen/Qwen2.5-1.5B-Instruct
        layer_range: [0, 28]
      - model: Qwen/Qwen2.5-Coder-1.5B-Instruct
        layer_range: [0, 28]
merge_method: slerp
base_model: Qwen/Qwen2.5-1.5B-Instruct
parameters:
  t: 0.5
dtype: bfloat16
'''
with open('merge_config.yaml', 'w') as f:
    f.write(config)
print('✅ 설정 저장 — 일반 Qwen(대화) + Coder Qwen(코딩)을 반반 섞는다')
```

```python
# mergekit을 실행하여 두 뇌를 합칩니다.
# 결과물은 'merged_model_output' 폴더에 저장됩니다.
!mergekit-yaml config.yaml ./merged_model_output

print("🎉 AI 두 개 합치기 수술 완료!")
```

```python
# 명령어 인식 문제를 피하기 위해 python -m 방식을 사용합니다.
# config.yaml 대신 bGJW1b6bUM2l 셀에서 만든 merge_config.yaml을 사용합니다.
!python -m mergekit.scripts.merge_yaml merge_config.yaml ./merged_model_output --allow-crimes --copy-tokenizer

print("🎉 모델 병합이 완료되었습니다!")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
import os

model_path = './merged_model_output'

if not os.path.exists(model_path):
    print(f"❌ 오류: '{model_path}' 폴더를 찾을 수 없습니다. 모델 병합 셀이 성공적으로 끝났는지 확인해 주세요.")
else:
    tok = AutoTokenizer.from_pretrained(model_path)
    model = AutoModelForCausalLM.from_pretrained(model_path, torch_dtype=torch.bfloat16, device_map='auto')

    def ask(q):
        msgs = [{'role':'user','content':q}]
        inputs = tok.apply_chat_template(msgs, add_generation_prompt=True, return_tensors='pt').to(model.device)
        out = model.generate(inputs, max_new_tokens=200, temperature=0.7, do_sample=True)
        print(tok.decode(out[0][inputs.shape[-1]:], skip_special_tokens=True))

    print('=== 대화 능력 ==='); ask('1인 기업가에게 좋은 아침 루틴 3가지 추천해줘')
    print('\n=== 코딩 능력 ==='); ask('파이썬으로 피보나치 수열 함수 짧게 써줘')
```
